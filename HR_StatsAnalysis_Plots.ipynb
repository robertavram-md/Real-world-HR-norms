```python
def clean_mc_dataset(dataset):
    dataset['hbp'] = dataset['hbp'].replace([2,3],[0,0],regex=True)
    dataset['high_chol'] = dataset['high_chol'].replace([2,3],[0,0],regex=True)
    dataset['diabetes'] = dataset['diabetes'].replace([2,3],[0,0],regex=True)
    dataset['blockages_in_your_coronary'] = dataset['blockages_in_your_coronary'].replace([2,3],[0,0],regex=True)
    dataset['heart_attack'] = dataset['heart_attack'].replace([2,3],[0,0],regex=True)
    dataset['pvd'] = dataset['pvd'].replace([2,3],[0,0],regex=True)
    dataset['clots'] = dataset['clots'].replace([2,3],[0,0],regex=True)
    dataset['chf'] = dataset['chf'].replace([2,3],[0,0],regex=True)
    dataset['stroke'] = dataset['stroke'].replace([2,3],[0,0],regex=True)
    dataset['arrhythmia'] = dataset['arrhythmia'].replace([2,3],[0,0],regex=True)
    dataset['copd'] = dataset['copd'].replace([2,3],[0,0],regex=True)
    dataset['asthma'] = dataset['asthma'].replace([2,3],[0,0],regex=True)
    #dataset['sex'] = dataset['sex'].replace([2,1],[1,0],regex=True)
    dataset['sleep_apnea'] = dataset['sleep_apnea'].replace([2,3],[0,0],regex=True)
    #dataset['new_race'] = dataset['new_race'].replace([2,1],[1,2],regex=True)
    dataset['disease_sum'] = dataset['hbp'] + dataset['high_chol'] + dataset['diabetes'] + dataset['heart_attack'] + dataset['chf'] + dataset['stroke'] + dataset['copd'] + dataset['asthma'] + dataset['sleep_apnea'] + dataset['arrhythmia'] 


    return dataset
```

```python
def clean_demographics(demographics):
    #1     : White
    #2       : Black or african AMerican
    #3      : Asian
    #4       : Other, do not disclose
    #8      : Hispanic
    #9      : Multiethnic
    
    #For q_hispanic:
   #Don't know                                                6.0
   #No                                                        1.0
   # Yes: Cuban                                                4.0
   # Yes: Mexican, Mexican American or Chicano                 2.0
   # Yes: Other or Mixed Hispanic, Latino or Spanish origin    5.0
   # Yes: Puerto Rican                                         3.0
   # Name: a_ref_id, dtype: float64

    df3 = demographics['race'].str.split('|', expand=True)
    df3.columns = ['race{}'.format(x+1) for x in df3.columns]
    df3['user_id'] = demographics['user_id']
    df3 = df3.fillna(0)
    
    values = {'hispanic': 1}
    demographics = demographics.fillna(value=values)
    try:
        df4 = demographics['hispanic'].str.split('|', expand=True)
        df4.columns = ['hispanic{}'.format(x+1) for x in df4.columns]
        df4['user_id'] = demographics['user_id']
        df4 = df4.fillna(0)
        df4['hispanic1'] = df4['hispanic1'].astype(int)

        display(df4['hispanic1'].value_counts())

        #Add multiethnic and hispanic_race as 'true or false' (BOOL) variable to demographics

        demographics['hispanic_race'] = np.where((df4['hispanic1'] == 1), 0,
                        np.where((df4['hispanic1'] != 1) & (df4['hispanic1'] != 6) & (df4['hispanic1'] != 0), 1, pd.np.nan))
    except:
        demographics['hispanic_race'] = np.where((demographics['hispanic'] == 1), 0,
                        np.where((demographics['hispanic'] != 1) & (demographics['hispanic'] != 6) &  (demographics['hispanic1'] != 0), 1, pd.np.nan))
    
    try:
        demographics['sex'] = demographics['sex'].astype(str).str.split('|').str[0]
        demographics['sex'] = pd.to_numeric(demographics['sex'], errors='ignore').astype(float, errors='ignore')
    except:
        print("No multiple sexes found")
    demographics['multiethnic'] = np.where((df3['race2']!=0), 1,
                np.where((df3['race2'] == 0), 0, pd.np.nan))
    #Either make the new race column as hispanic_race, multiethnic or the value in race1
    demographics['new_race'] = np.where((demographics['hispanic_race'] == 1), 8,
                    np.where((demographics['multiethnic'] == 1) & (demographics['hispanic_race'] == 0),  9, df3['race1']))
    demographics['new_race'] = demographics['new_race'].astype(int)
    demographics['new_race'] = demographics['new_race'].replace([0,5,6,7], 4)
    demographics['new_race'] = demographics['new_race'].replace([2,1],[1,2],regex=True)
    demographics['new_race'] = pd.to_numeric(demographics['new_race'], errors='coerce')
    
    return demographics
```

```python
def compute_norms_according_to_age(azumio_age, output):
    # Arrange the data in strata of 5 & 10

    labels_values =  [ "{0} - {1}".format(i, i + 10) for i in range(30, 110, 10) ]
    labels_age = [ "{0} - {1}".format(i, i + 10) for i in range(10, 100, 10) ]

    df_age_category = pd.DataFrame(azumio_age, columns=['geomean_bpm', 'age', 'n_bpm', 'user_id'])

    #use pd.cut function can attribute the values into its specific bins
    df_age_category['value_cat'] = pd.cut(azumio_age.geomean_bpm, range(30, 115, 10), right=True, labels=labels_values)
    df_age_category['age_cat'] = pd.cut(azumio_age['age'], range(10, 108, 10), right=True, labels=labels_age)
    display(df_age_category.head(n=5))
    df = df_age_category.groupby('age_cat')['user_id'].nunique()
    display(df.head(n=10))
    df = df_age_category.groupby('age_cat')['n_bpm'].agg(f)
    display(df.head(n=10))

    df = df_age_category.groupby('age_cat')['n_bpm'].agg(f)
    display(df.head(n=10))
    print("Age cat mean")
    df = df_age_category.groupby('age_cat').mean()
    display(df.head(n=10))
    print("Standard deviation of age cat")
    df = df_age_category.groupby('age_cat').std()
    display(df.head(n=10))
    #display(df_age_category.loc[df_age_category['age_cat'] == "70 - 80"])
    print(stats.ttest_ind(df_age_category['geomean_bpm'].loc[df_age_category['age_cat'] == "10 - 20"], df_age_category['geomean_bpm'].loc[df_age_category['age_cat'] == "60 - 70"]))
    display(np.percentile(df_age_category['geomean_bpm'],97.5))
    display(np.percentile(df_age_category['geomean_bpm'],2.5))
    sns.set_style("whitegrid")
    dez = pd.DataFrame(columns=('age', '95p', '90p', '75p', '50p', '25p', '10p', '5p'))
    i = 0 

    for age, value in df_age_category.groupby(['age']):
        new_entry = {'age': age, '95p': np.percentile(value['geomean_bpm'],95), '90p': np.percentile(value['geomean_bpm'],90), '75p': np.percentile(value['geomean_bpm'],75), '50p': np.percentile(value['geomean_bpm'],50), '25p': np.percentile(value['geomean_bpm'],25), '10p': np.percentile(value['geomean_bpm'],10), '5p': np.percentile(value['geomean_bpm'],5)}
        dez.loc[len(dez)] = new_entry
        i = i+1
        if i == 90: 
            break


    plt.ylim(50, 130)
    plt.xlim(18, 80)
    sns.set(font_scale=1.5)
    ax = sns.regplot(x="age", y="95p", data=dez,  x_estimator=np.median, lowess=True, truncate=True, scatter_kws={"s": 0}, dropna=True, label='95p', line_kws={"color": "darkred"} )
    ax = sns.regplot(x="age", y="90p", data=dez,  x_estimator=np.median, lowess=True, truncate=True, scatter_kws={"s": 0}, dropna=True, label='90p', line_kws={"color": "darkgreen"})
    ax = sns.regplot(x="age", y="75p", data=dez,  x_estimator=np.median, lowess=True, truncate=True, scatter_kws={"s": 0}, dropna=True, label='75p', line_kws={"color": "darkblue"})
    ax = sns.regplot(x="age", y="50p", data=dez,  x_estimator=np.median, lowess=True, truncate=True, scatter_kws={"s": 0}, dropna=True, label='50p', line_kws={"color": "black", "lw":6})
    ax = sns.regplot(x="age", y="25p", data=dez,  x_estimator=np.median, lowess=True, truncate=True, scatter_kws={"s": 0}, dropna=True, label='25p', line_kws={"color": "brown"})
    ax = sns.regplot(x="age", y="10p", data=dez,  x_estimator=np.median, lowess=True, truncate=True, scatter_kws={"s": 0}, dropna=True, label='10p', line_kws={"color": "orange"})
    ax = sns.regplot(x="age", y="5p", data=dez,  x_estimator=np.median, lowess=True, truncate=True, scatter_kws={"s": 0}, dropna=True, label='5p', line_kws={"color": "green"})
    ax.figure.set_size_inches(18.5, 15.5)
    plt.ylabel("Heart rate (BPM)", fontsize=24)
    # Set y-axis label
    plt.xlabel("Age (y)", fontsize=24)
    sns.despine()

    dez.reset_index().to_csv(output)
    return 'True'
```
